```python
import pandas as pd
```

```python
df = pd.read_csv("character_nicknames_cleaned.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
df = df.dropna(subset=['nickname'])
```

```python
df = df.drop_duplicates()
```

```python
df['nickname'] = df['nickname'].str.strip()
```

```python
df['nickname'] = df['nickname'].str.title()  # e.g., 'last pharaoh' â†’ 'Last Pharaoh'
```

```python
df.to_csv("character_nicknames_cleaned.csv", index=False)
```
